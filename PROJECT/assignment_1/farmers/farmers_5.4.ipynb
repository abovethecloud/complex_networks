```python
import networkx as nx  # to work on the graph
import matplotlib.pyplot as plt
import dzcnapy_plotlib as dzcnapy
```

```python
G = nx.Graph()
```

setting the parameters of the graph

```python
n_farmers = 50
#friendship_radius_proportion = 2/5
#friendship_radius = n_farmers * friendship_radius_proportion
friendship_radius = 20
```

```python
for i in range(0, n_farmers):
    attr = 'Farmer '+ str(i)
    G.add_node(i, name= attr)
    min = i-friendship_radius if (i-friendship_radius) >= 0 else 0
    G.add_weighted_edges_from([(i, a, 1) for a in range(min, i)])
    if min > 0:
        G.add_weighted_edges_from([(i, b, -1) for b in range(0, min)])
```

```python
print(nx.info(G))
```

```python
print(G.edges(data = True))
```

```python
nx.draw_networkx(G)
```

```python
nx.write_graphml(G, 'farmers.graphml', prettyprint = True)
```

```python
pos = nx.spectral_layout(G)  # There are different types of layout
nx.draw_networkx(G, pos=pos, with_labels=True,**dzcnapy.medium_attrs)
```
